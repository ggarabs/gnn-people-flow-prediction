```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [2015, 2016, 2017, 2018, 2019, 2020]
y = [1, 2, 4, 6, 8, 9]

df = pd.DataFrame({'Ano':x, 'Smart Campus': y})
```

```python
plt.figure(figsize=(9, 4))
sns.lineplot(data=df, x='Ano', y='Smart Campus', marker='o')
plt.title('NÃºmero de Smart Campus no Brasil por Ano')
plt.xlabel('Ano')
plt.ylabel('Smart Campus')
plt.savefig('grafico.svg', bbox_inches='tight')
plt.show()
```
